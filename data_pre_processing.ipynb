 The code chunk below encodes the counter_type column and deltes the counter_statue column.

```python
# Transform 'counter_type' feature to a numeric variable (example: using LabelEncoder)
from sklearn.preprocessing import LabelEncoder

STEG_BILLING_HISTORY['counter_type_encoded'] = LabelEncoder().fit_transform(STEG_BILLING_HISTORY['counter_type'])

# Delete the 'counter_statue' feature
STEG_BILLING_HISTORY = STEG_BILLING_HISTORY.drop('counter_statue', axis=1)

STEG_BILLING_HISTORY.head()
```
